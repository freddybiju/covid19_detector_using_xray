```python
import pandas as pd
import os #creating folders
import shutil #AND MOVING FILES
```

```python
#create the data for positive samples(github)
FILE_PATH = "C:/Users/fredd/Desktop/jupyterproject/metadata.csv"
IMAGES_PATH = "C:/Users/fredd/Desktop/jupyterproject/images"

```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head(40)
```

```python
TARGET_DIR = "C:/Users/fredd/Desktop/jupyterproject/dataset/covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("covid folder created") #creating folders using os
```

```python
cnt = 0  #counting numnber of covid postive pictures and we want only front view
for(i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA": #only  covid 19 and front side
        filename = row["filename"] #accessing row filename 
        image_path = os.path.join(IMAGES_PATH,filename)  #combining filename with image path
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)  #copying
        print("moving image",cnt)  
        cnt+=1
print(cnt)
```

```python
#sampling of images from kaggle
import random
KAGGLE_FILE_PATH = "C:/Users/fredd/Desktop/jupyterproject/NORMAL"
TARGET_NORMAL_DIR = "C:/Users/fredd/Desktop/jupyterproject/dataset/normal"
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH) #getting image names
```

```python
image_names
```

```python
random.shuffle(image_names) #shuffling images
```

```python
for i in range(196): # copy starting 196 pics which were randomized
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,image_name)
    
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("copying image",i)
```
